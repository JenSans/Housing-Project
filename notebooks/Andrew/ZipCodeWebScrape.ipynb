```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.ciclt.net/sn/clt/capitolimpact/gw_ziplist.aspx?FIPS=53033'

res = requests.get(url)
res.status_code
```

```python
soup = BeautifulSoup(res.content)

tables = soup.find('table')
```

```python
list(tables.children)
```

```python
tablea = list(tables.children)[1]
tableb = tablea.find('table')
tablec = list(list(tableb.children)[0].children)
tablec
```

```python
table3 = [tablec[index] for index in [1, 3, 5]]

all_tables = []

for table in table3:
    for row in table.find_all('tr')[1:]:
        for entry in row.find_all('td'):
            all_tables.append(entry.text)
```

```python
all_tables
```

```python
zips = all_tables[::3]
cities = all_tables[1::3]

zips = pd.DataFrame(zips, columns=['ZipCode'])
cities = pd.DataFrame(cities, columns=['City'])

df = pd.concat([zips, cities], axis=1)
df.head()
```
